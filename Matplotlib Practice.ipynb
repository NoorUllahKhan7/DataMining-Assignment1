##  Making plots using Numpy

```python
#importing numpy
import numpy as np

#Importing matlotlib

import matplotlib.pyplot as plt
#creating numpy array
x = np.linspace(0,10,100)
```

```python
#Creating a subplot using matplotlib
fig, ax = plt.subplots()

#For plotting line plot using numpy array

ax.plot(x, x**2);
```

```python
# plotting a scatter plot using numpy array

fig, ax = plt.subplots()

ax.scatter(x, np.sin(x));

# putting some explanation on axix 

ax.set(title="scatter plot of sin wave", xlabel="x axis",ylabel="y axis");
```

```python
# Plotting scatter plot for a dictionary

butter_shop = {"Almond butter":4,
              "Apricot butter":10,
              "Butter butter":10 }

fig,axs = plt.subplots()

axs.bar(butter_shop.keys(), butter_shop.values());

axs.set(title="Niazi butter shop" );
```

```python
#Creating histogram

fig , axs = plt.subplots()
axs.hist(np.random.rand(1000));
```

```python
# Creating four subplots in a figure
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows= 2,
                                         ncols= 2, figsize=(10,10))

ax1.plot(x,x**2);
ax2.scatter(x,x**2);
ax3.hist(np.random.random(1000));
ax4.bar(butter_shop.keys(),butter_shop.values());


```

# Making plots using object oriented Matplotlib

```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
#importing pandas to rad csv
import pandas as pd

#importing heart disease csv file 
heart_dis = pd.read_csv("heart-disease.csv")
```

```python
heart_dis.head()
```

```python
#we want to use data of people of age over 50 years 
over_50 = heart_dis[heart_dis["age"] > 50]
over_50.head()
```

```python
#Making subplot out of this dataset for cholestrol level 
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]
                     )
ax.set(title="Heart disease and cholestrol levels", xlabel="age", ylabel="cholestrol");

# we will add a legend by using target column which is refered in scatted method as c argument
ax.legend(*scatter.legend_elements(),title="Target");


# Add a horizontal line in plot
ax.axhline(over_50["chol"].mean() , linestyle="--");
```

```python
#Creating two subplots in a figure using Object Oriented matplotlib

fig, (ax0,ax1) = plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                            sharex= True)
#First subplot
scatter1 = ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"])
ax0.set(title="Heart disease and Cholestrol",
       xlabel="age",
       ylabel="cholestrol");
ax0.axhline(over_50["chol"].mean(), linestyle="--")
#Addingg legend
ax0.legend(*scatter1.legend_elements(), title="Target")
#Second Subplot

scatter2 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

ax1.set(title="Heart Disease and max heart rate",
       xlabel="age",
       ylabel="Max heart rate");
ax1.axhline(over_50["thalach"].mean() , linestyle="--")
ax1.legend(*scatter2.legend_elements(), title="Target")
#Adding a title to the whole figure be like
fig.suptitle("Heart disease Analysis",fontsize=16,fontweight="bold");
                             
```
